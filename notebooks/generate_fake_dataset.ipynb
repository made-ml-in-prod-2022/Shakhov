```python
import pandas as pd
import numpy as np
```

```python
real_data = pd.read_csv('../data/raw/heart_cleveland_upload.csv')
real_data.head()
```

```python
def generate_fake_dataset(real_data : pd.DataFrame, n_rows : int) -> pd.DataFrame:
    fake_data = real_data[0:n_rows].copy()
    for i in range(n_rows):
        for j in real_data.columns:
            fake_data.at[i, j] = real_data[j][np.random.randint(0, real_data.shape[0])]

    return fake_data
```

```python
fake_data = generate_fake_dataset(real_data, 100)
```
